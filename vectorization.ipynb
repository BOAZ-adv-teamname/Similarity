```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data= pd.read_csv('./Sum_Database/summary_gpu.csv')
```

```python
corpus = [i for i in data.iloc[:10,0]]
```

```python
corpus
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
print(X.shape)
```

```python
x1=X[-1]
x2=X[:-1]
```

```python
result = cosine_similarity(x1,x2)[0]
```

```python
result2 = np.sort(result)[::-1]
```
